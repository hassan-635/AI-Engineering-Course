# Properties of Matrix Multiplication Verification :

### 1. AB is not equal to BA

```python
A = np.array(([1, 2], [3, 4]))
B = np.array(([5, 6], [7, 8]))
AB = np.dot(A, B)
BA = np.dot(B, A)
```

```python
print(AB)
```

```python
print(BA)
```

### 3. Associative property (AB)C = A(BC)

```python
C = np.array(([9, 10], [11, 12]))
AB = np.dot(A, B)
BC = np.dot(B, C)
lhs = np.dot(AB, C)
rhs = np.dot(A, BC)
```

```python
print(lhs)
```

```python
print(rhs)
```

### 3. Distributive Property A(B+C) = AB + AC

```python
b_plus_c = B + C
lhs = np.dot(A, b_plus_c)
AC = np.dot(A, C)
rhs = AB + AC
```

```python
lhs
```

```python
rhs
```
