# Matrices :

```python
import numpy as np
```

```python
A = np.array(([1, 2], [3, 4]))

B = np.array(([5, 6], [7, 8]))
```

```python
shape1 = A.shape
dim1 = A.ndim
shape2 = B.shape
dim2 = B.ndim
```

```python
print("Shape : ")
print(shape1)
print(shape2)
```

```python
print("Dimensions : ")
print(dim1)
print(dim2)
```

```python
transpose1 = A.T
transpose2 = B.T

print("Transpose : ")
print(transpose1)
print(transpose2)
```

```python
det1 = np.linalg.det(A)
det2 = np.linalg.det(B)

print("Determinant : ")
print(det1)
print(det2)
```

```python
inv1 = np.linalg.inv(A)
inv2 = np.linalg.inv(B)

print("Inverse : ")
print(inv1)
print(inv2)
```

```python
addition = A + B
subtraction = A - B
multiplication = np.dot(A, B)
mul = A * B

print(addition)
print(subtraction)
print(multiplication)
print(mul)
```

```python
block1 = np.array(([2, 1], [3, 4]))
block2 = np.array(([5, 6], [7, 8]))

blk = np.block([
    [block1, np.zeros((2, 3), dtype=int)],
    [np.zeros((1, 2), dtype=int),np.array([[9]]), np.zeros((1, 2), dtype=int)],
    [np.zeros((2, 3), dtype=int), block2]
])

print(blk)
```

```python
det = np.linalg.det(A)
print(det)
```

```python
inverse = np.linalg.inv(A)
print(inverse)
```

```python
m = np.dot(A, inverse)
print(m)
```

```python
iden = np.eye(5)
print(iden)
zero = np.zeros((5, 5))
print(zero)
one = np.ones((5, 5))
print(one)
```

```python
mat = iden + blk
print(mat)
```
