# Matrices :

```python
import numpy as np
```

```python
A = np.array(([1, 2], [3, 4]))

B = np.array(([5, 6], [7, 8]))
```

```python
shape1 = A.shape
dim1 = A.ndim
shape2 = B.shape
dim2 = B.ndim
```

```python
print("Shape : ")
print(shape1)
print(shape2)
```

```python
print("Dimensions : ")
print(dim1)
print(dim2)
```

```python
transpose1 = A.T
transpose2 = B.T

print("Transpose : ")
print(transpose1)
print(transpose2)
```
